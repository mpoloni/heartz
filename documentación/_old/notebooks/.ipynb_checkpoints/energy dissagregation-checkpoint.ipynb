# Energy Dissagregation #

Estima el conusmo individual de los equipos a partir del consumo agregado. Se debe usar formato REDD

```python
import nilmtk
from nilmtk import *
from nilmtk import DataSet, TimeFrame, MeterGroup, HDFDataStore
from nilmtk.utils import print_dict
from nilmtk.disaggregate import CombinatorialOptimisation, fhmm_exact

from pylab import rcParams
import matplotlib.pyplot as plt
%matplotlib inline
rcParams['figure.figsize'] = (14, 6)
plt.style.use('ggplot')

import numpy as np
import pandas as pd
from os.path import join
```

```python
ds = DataSet('/home/matias/Descargas/Heartz/one_house/',format='CSV')
```

```python
print_dict(ds.buildings)
```

```python
print_dict(ds.metadata)
```

```python
building = ds.buildings[1]
elec = building.elec
print(elec)
```

```python
elec.appliances
```

```python
fraction = building.elec.submeters().fraction_per_meter().dropna()
```

```python
labels = elec.get_labels(fraction.index)
plt.figure(figsize=(10,10))
fraction.plot(kind='pie', labels=labels);
```

```python
co = CombinatorialOptimisation()
co_train = co.train(elec)
```

```python
disag_filename = join('/home/matias/Descargas/Heartz/one_house/', 'disag-co.h5')
output = HDFDataStore(disag_filename, 'w')
co.disaggregate(elec.mains(), output)
output.close()
```
